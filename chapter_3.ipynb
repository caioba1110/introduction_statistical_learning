Existe um ponto interessenta neste modelo. Se tivermos duas variáveis A e B. Sendo A e B correlacionadas, porém pode ser que elas não tragam algum beneficio para o modelo, afinal, suponhamos somente que A seja benéfica para o modelo. Quando A subir, B também irá subir como também y irá subir, portanto, B mesmo não sendo benéfica ao modelo, irá parecer que B é boa para predizer y. Porém não. Existe um ótimo exemplo para mostrar isso, quando ocorre ataques de tubarão provavelmente também terá uma alta venda de sorvetes no dia. A venda de sorvetes é correlacionada com o ataque de tubarão? Sim, mas não exatamente. É uma falsa correlação.

Uma forma de provar isso é fazendo um teste de hipotese para comprovar está ideia.

Como podemos saber qual variável colocar em nosso modelo?

Bom, podemos fazer o teste clássico de testar com e sem a variável, todavia, caso você tenha muitas variáveis, é inviável fazer isso.

Temos também o método foward selecion, nele treinamos p modelos de regressão linear simples, pegamos a variavel do modelo com menor RSS. Então refazemos tudo isso de novo com duas variáveis (a que você acabou de pegar e com as outras p) p-1 vezes, e vai fazendo isso até alguma regra o faça parar (rss muito grande, muitas variáveis, enfim).

Podemos também fazendo backward selection, que consiste em realizarmos um treino de modelo com todas as variáveis e com isso remover a variável com maior p-value. Depois refaz o treino (sem aquela variavel) e assim vai, até alguma regra o parar, como por exemplo, o maior p-value ser pequeno de mais.
