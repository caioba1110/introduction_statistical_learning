## MSE - mean squared error

Quando queremos treinar um modelo, necessariamente precisamos fazer testes para sabermos quão bom eles estão.
Um dos testes mais comuns em regressão é o mean squared error (MSE)
Dado pôr:

Quanto menor der o MSE, mais próximos as predições estarão dos dados reais.

Quando treinamos, nós temos dados de treino(para treinar) e de teste(para testar).
Se usarmos os dados de treino para teste (usando MSE), não iremos ter necessáriamente dados confiaveis, como mostra o gráfico a seguir. 

No gráfico a direita, podemos ver que a linha cinza mostra que quanto mais flexiveis, mais os dados de treino terão um MSE menor. Todavia a linha vermelha que são os dados de teste sobem muito conforme o modelo fica flexivel.

Quando o MSE de treino diminui e o MSE de teste aumenta, podemos dizer que ocorreu um overfit. Isso acontece porquê o modelo está tentando encontrar padrões onde não existem e simplesmente aconteceu algo aleatório. Quando overfitamos o modelo, normalmente os dados de teste serão gigantes, afinal, ele tentará encontrar padrões onde não existem.

Está equação define o MSE de teste esperado. Está equação nos diz que quando quisermos diminuir o MSE, precisamos de um método estátistico que diminua tanto a variancia quanto o bias (viés).

Existem um ponto interessante, quanto mais aumentarmos a flexibilidade do modelo maior vai ser a variancia, todavia, menor vai ser o viés. A principio o viés diminuindo, diminui muito o MSE esperado, mesmo aumentando um pouco a variancia, porém conforme vai ficando flexivel, a variancia aumenta muito, causando um aumento do MSE. Ou seja, existe um ponto perfeito de flexibilidade, nem muito, nem pouco.
