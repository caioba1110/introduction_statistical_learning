## Linear Discriminant Analysis (LDA)

O LDA assumo um valor muito melhor que uma RL principalmente quando:

    * Quando as classes estão muito bem separadas
    
    * Se n for pequeno e a distribuição de X for aproximadamente normal em cada uma das classes
    
    * LDA também é muito popular quando queremos prever mais de uma classe.

O objetivo do LDA é maximizar a distancia entre as classes e diminuir a variancia dos exemplos de cada classe. Como no exemplo:



Uma das formas de fazer isso é usar o teorema de bayes


(OBS: Para o correto funcionamento do treino, devemos assumir que as variáveis explicativas sejam normais e a inexistencia de outliers)


Por padrão, devesse usar no minimo 20 amostras para cada variavel explicativa, outro ponto importante é que as classes devem ter a mesma quantidade de amostras, afinal, não se deve dar mais importancia para uma variavel do que outra.

```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
y = np.array([1, 1, 1, 2, 2, 2])
clf = LinearDiscriminantAnalysis()
clf.fit(X, y)

print(clf.predict([[-0.8, -1]]))
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import numpy as np
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
y = np.array([1, 1, 1, 2, 2, 2])
clf = QuadraticDiscriminantAnalysis()
clf.fit(X, y)

print(clf.predict([[-0.8, -1]]))
```
